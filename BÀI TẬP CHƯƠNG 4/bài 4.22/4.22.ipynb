```python
stocks1['high']=stocks1['high'].fillna(stocks1['high'].max())
stocks1.head()
```

```python
stocks1['low']=stocks1['low'].fillna(stocks1['low'].min())
stocks1.head() 
```

# 3

```python
stocks = pd.concat((stocks1,stocks2),axis=0) 
stocks.head()
```

```python
stocks_company = pd.concat((stocks,companies),axis = 1)
stocks_company.head()
```

# 5

```python
print("giá open, high, low, close trung bình ở mỗi công ty là: ")
stocks_company.groupby('symbol')[['open','high','low','close']].mean()
```

# 6

```python
print('giá đóng cửa trung bình,lớn nhát,nhỏ nhất của mỗi công ty là :')
stocks_company.groupby('symbol').agg({'close': ['mean','max','min']})
```
