# Lab3.2

```python
# Kiểm tra dữ liệu null nào k
if stocks1.isna().any().any():
    print("stocks1 co' null")
else:
    print("stocks1 kh co null")
```

```python
# Thay thế dữ liệu null
stocks1['high'] = stocks1['high'].fillna(stocks1['high'].mean())
stocks1['high']
```

```python
# Thay thế null vào cột low
stocks1['low'] = stocks1['low'].fillna(stocks1['low'].mean())
stocks1['low']
```

```python
# Đọc file stock2
file2 = r'D:\17A1KHDL\LAB3\data\stocks2.csv'
stocks2 = pd.read_csv(file2)
stocks2.head()
```

```python
# Gộp stock1+2
stocks = pd.concat([stocks1,stocks2],axis=0)
stocks
```

```python
 # Tạp multiindex
stocks = stocks.set_index(['date','symbol'],)
stocks.head()
```

```python
 # tạo pivort
pivort_table = pd.pivot_table(stocks,index='date',columns='symbol',values='close',aggfunc='mean')
pivort_table
```

```python
# Thêm cộht
pivort_table.loc['Total volume'] = stocks.groupby('symbol')['volume'].sum()
```

```python
pivort_table
```

```python
# sắp xép từ thấp đến cao
pivort_table = pivort_table.sort_values(by='Total volume',axis=1,ascending=False)
```
