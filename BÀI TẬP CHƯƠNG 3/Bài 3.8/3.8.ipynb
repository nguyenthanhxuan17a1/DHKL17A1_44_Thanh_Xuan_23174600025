```python
import numpy as np 
```

```python
np_posistion = np.loadtxt('D:\\BÀI TẬP\\BÀI TẬP CHƯƠNG 3\\Bài 3.8\\positions.txt',delimiter=',',encoding='utf-8-sig',dtype=str)
np_posistion
```

```python
np_height = np.loadtxt('D:\\BÀI TẬP\\BÀI TẬP CHƯƠNG 3\\Bài 3.8\\heights.txt',delimiter=',',encoding='utf-8-sig',dtype=float)
np_height
```

```python
position = np.reshape(np_posistion,(-1,1))
heights = np.reshape(np_height,(-1,1))
```

```python
arr = np.hstack((position,heights))
arr
```

```python
mask = arr[:,0] =='GK'
arr_gk =arr[mask]
```

```python
arr_height_gk = arr_gk[:,1].astype(float)
arr_height_gk
```

```python
# chiều cao trung bình gk là:
np.mean(arr_height_gk)
```

```python
#3
mask1 = arr[:,0] != 'GK'
arr_diffrent_gk = arr[mask1]
arr_diffrent_gk
```

```python
height_diffrent_gk = arr_diffrent_gk[:,1].astype(float)
height_diffrent_gk
```

```python
# chiều cao trung bình kh phải GK là
np.mean(height_diffrent_gk)
```
